```python
# Import Modules
import pandas as pd
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
# Importing dataset
df = pd.read_csv('data/xyz.csv')
```

# S&P500 Futures Volatility Surface
Using the data from this [model](https://github.com/antonio-hickey/Finance/blob/main/Volatility%20Surface/data/model.py) we wil render a 3d surface of implied volatility of ES options. 

We can use this surface to better understand how the market is pricing volatility.

### Dimensions:

    1. (x) Strike
    2. (y) Days Until Expiration
    3. (z) Implied Volatility

```python
# Defining Data
x = df["x"].values
y = df["y"].values
zCall = df["zCall"].values
zPut = df["zPut"].values

# Reshaping data
x = np.reshape(x,(6,10))
y = np.reshape(y,(6,10))
zCall = np.reshape(zCall,(6,10))
zPut = np.reshape(zPut,(6,10))

# Render Figure
fig = plt.figure(figsize=(14,9))
ax = plt.axes(projection='3d')
my_cmap = plt.get_cmap('cividis')
ax.plot_surface(x,y,zCall,cmap=my_cmap)
ax.set_xlabel("Strikes")
ax.set_ylabel("Days Until Expiration")
ax.set_zlabel("Implied Volatility")
plt.title("Volatlity Surface For Calls")
```

```python
# Render Figure
fig = plt.figure(figsize=(14,9))
ax = plt.axes(projection='3d')
my_cmap = plt.get_cmap('cividis')
ax.plot_surface(x,y,zPut,cmap=my_cmap)
ax.set_xlabel("Strikes")
ax.set_ylabel("Days Until Expiration")
ax.set_zlabel("Implied Volatility")
plt.title("Volatlity Surface For Puts")
```
